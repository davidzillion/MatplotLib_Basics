# MatplotLib Basics

## Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## General information to know

From the module to which we have renamed plt:

* We will use the `.title()` method to add a title to the plot.
* We will use the `.xlabel()` method to add a label to the horizontal axis of the 2D plot.
* We will use the `.ylabel()` method to add a label to the vertical axis of the 2D plot.
* We will use the `.legend()` method to display a legend.

To change the size of the plot we can use `.subplots()`. Example: `fig, ax = plt.subplots(figsize=(15, 9))`.

## Plot types

### Scatter Plot

To make a point cloud graph, we use the `.scatter ()` method of the `plt` module.

Some of the parameters of this method are:

* `x`: scalar, array or list that indicates the first coordinate of the observations
* `y`: scalar, array or list that indicates the second coordinate of the observations
* `c`: to change the fill color
* `edgecolors`: to change the color of the outline
* `alpha`: to change the transparency
* `marker`: to change the [point shape](https://matplotlib.org/3.2.1/api/markers_api.html#module-matplotlib.markers)
* `s`: to change the size of the points (it is measured in points)
* `linewidths`: to change the thickness of the outline

```python
height = [174.3, 153.5, 162.1, 157.9, 174.8, 169.2, 172.0, 160.4, 152.8, 163.3] 
weight = [ 65.7,  59.2,  57.3,  61.5,  77.3,  90.7,  85.4,  63.1,  73.2, 105.5]

title = "Height vs. weight"
xlabel = "height (cm)"
ylabel = "weight (kg)"

fig, ax = plt.subplots(figsize=(15, 9))

plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.scatter(x = height, y = weight, 
            c = "black", edgecolors = "red", alpha = 1,
            marker = "X", s = 200, linewidths = 2)
plt.show()
```

**Attenbtion** We could show 2 different sets of data in a point cloud graph:

```python
height_boys = [174.3, 153.5, 162.1, 157.9, 174.8, 169.2, 172.0, 160.4, 152.8, 163.3] 
weight_boys = [65.7, 59.2, 57.3, 61.5, 77.3, 90.7, 85.4, 63.1, 73.2, 105.5]

height_girls = [154.2, 156.1, 160.3, 157.9, 162.6, 153.9, 170.1, 165.2, 157.6, 163.5] 
weight_girls = [55.3, 54.2, 47.3, 62.4, 77.5, 60.3, 52.4, 58.1, 50.2, 80.1]

plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)

plt.scatter(x = height_boys, y = weight_boys, c = "red", 
            marker = "^", s = 100)
plt.scatter(x = height_girls, y = weight_girls, c = "blue", 
            marker = "v", s = 100)

plt.show()
```

**Observation:** We can draw each point of a size, passing a list to the `s` parameter:

```python
x = [1, 2, 3, 4, 5, 6, 7]
sizes = [100, 150, 200, 250, 300, 350, 400]

plt.scatter(x = x, y = x, s = sizes, c = "m", alpha = 0.5)
plt.title("Different sizes")
plt.show()
```

**Observation:** We can paint a point of each color, passing a list of colors as parameter `c`:

```python
x = [1, 2, 3, 4, 5, 6, 7]
colors = ["red", "orange", "yellow", "green", "cyan", "blue", "purple"]
sizes = [100, 150, 200, 250, 300, 350, 400]

plt.scatter(x = x, y = x, c = colors, s = sizes)
plt.title("Rainbow")
plt.show()
```

We could also have used a color map to color the dots.

To do this, instead of a list of colors, we will need to indicate in the `cmap` parameter the color map to use and, in the` c` parameter, indicate a list or 1D array of values between 0 and 100 to indicate the color of each point for the given colormap.

```python
colors = np.array([0, 10, 20, 30, 40, 50, 60])

plt.scatter(x = x, y = x, c = colors, cmap = "plasma", s = 200)
plt.title("Color map")
plt.show()
```

**Observation:** We can attach to the graph the color map used with the `.colorbar ()` method.

```python
colors = np.array([0, 10, 20, 30, 40, 50, 60])

plt.scatter(x = x, y = x, c = colors, cmap = "plasma", s = 200)
plt.title("Color Map")
plt.colorbar()
plt.show()
```

**Observation:** We can invert the color map by adding `_r` to the name of the color map

## Line Plots

To make a line plot, we use the `.plot ()` method of the `plt` module.

Some of the parameters of this method are:

* `x`: scalar, array or list that indicates the first coordinate of the observations.
* `y`: scalar, array or list that indicates the second coordinate of the observations
* `color`: to change the fill color. We can also refer to this parameter by its diminutive `c`
* `fmt`: to quickly set a basic format as string. For example, `" or "` are red circles. The recommended order for entering the format in this way is `" [marker] [linestyle] [color] "` although `" [color] [marker] [linestyle] "` is also supported.
* `linewidth`: to change the thickness of the line. We can also refer to this parameter by its diminutive `lw`
* `linestyle`: to change the style of the line. We can also refer to this parameter by its diminutive `ls`
* `alpha`: to change the transparency
* `marker`: to change the [point shape](https://matplotlib.org/3.2.1/api/markers_api.html#module-matplotlib.markers). If we do not indicate this parameter, the points are not drawn.
* `markersize`: to change the size of the points. We can also refer to this parameter by its diminutive `ms`
* `markeredgecolor`: to change the color of the point outline. We can also refer to this parameter by its diminutive `mec`
* `markerfacecolor`: to change the fill color of the point. We can also refer to this parameter by its diminutive `mfc`

For the `fmt` parameter, the options available for` marker`, `linestyle` and` color` are

| `marker` | shape | `linestyle` | line style | `color` | color |
| : ---: | : --- | : ---: | : --- | : ---: | : --- |
| `"."` | point | `"-"` or `"solid"` | solid | `"b"` o `"blue"` | blue |
| `","` | pixel | `"-"` or `"dashed"` | discontinuous | `"g"` or `"green"` | green |
| `"or"` | circle | `"-."` or `"dashdot"` | dash-dot | `"r"` or `"red"` | red |
| `"v"` | downward triangle | `":"` or `"dotted"` | points | `"c"` or `"cyan"` | cyan |
| `"^"` | triangle up | `"None"` | no line | `"m"` or `"magenta"` | magenta |
| `"<"` | triangle to the left | | | `"and"` or `"yellow"` | yellow |
| `">"` | right triangle | | | `"k"` or `"black"` | black |
| `"1"` | tri down | | | `"w"` or `"white"` | white |
| `"2"` | tri up | | | | |
| `"3"` | tri to the left | | | | |
| `"4"` | tri to the right | | | | |
| `"s"` | square | | | | |
| `"p"` | pentagon | | | | |
| `"*"` | star | | | | |
| `"h"` | hexagon 1 | | | | |
| `"H"` | hexagon 2 | | | | |
| `"+"` | cross | | | | |
| `"x"` | x | | | | |
| `"D"` | diamond | | | | |
| `"d"` | fine diamond | | | | |
| `"\|"` | vertical bar | | | | |
| `"_"` | horizontal bar | | | | |

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.plot(sorted(height), weight, 
            c = "blue", ls = "--", lw = 2,
            marker = "*", ms = 20, mfc = "m")
plt.show()
```

We could get the same result using the `fmt` parameter:

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.plot(sorted(height), weight, "*--b", lw = 2,
         markersize = 20, mfc = "m")
plt.show()
```

**Observation:** If we do not indicate parameter `x`, the integers 0, 1, 2, ..., $ n-1 $ are considered as first coordinates, where $n$ is the total number of observations:

```python
rainy_days = [20, 15, 19, 25, 13, 8, 3, 5, 10, 16, 20, 18]

plt.title("Rainy days by month")
# (0 = January, 1 = February, ..., 11 = December)
plt.xlabel("Month")
plt.ylabel("Total rainy days")
plt.plot(rainy_days, c = "cyan", marker = "o")
plt.show()
```

**Observation:** In the same plot we can draw more than one line:

```python
l1 = np.array([3, 5, 1, 9])
l2 = np.array([8, 2, 6, 4])

plt.plot(l1, c = "blue")
plt.plot(l2, c = "magenta")
plt.title("2 lines in 1 plot")
plt.show()
```

### Bar Plot

To make a bar chart, we use the `.bar()` method of the `plt` module.

Some of the parameters of this method are:

* `x`: float, array or list that indicates the categories or coordinates of the bars
* `height`: float, array or list that indicates the heights of the bars
* `width`: to change the width of the bars
* `linewidth`: to change the border size of the bars
* `bottom`: to change the minimum vertical value of the bars
* `align`: position of the horizontal axis mark with respect to the base of the bars. By default it is "center"
* `color` to change the colors of the bars. If we pass a list, we can assign a different color to each bar
* `edgecolor`: to change the color of the edge of the bars

```python
grades = ["Fail", "Approved", "remarkable", "Excellent"]
count = [35, 55, 23, 7]

plt.bar(grades, count, color = ["#ff7f7d", "#ffc07d", "#817dff", "#7dff86"], width = 0.5)
plt.xlabel("Marks")
plt.ylabel("Number of students")
plt.title("Marks of the students of one group")
plt.show()
```

```python
plt.barh(grades, count, color = ["#ff7f7d", "#ffc07d", "#817dff", "#7dff86"], height = 0.5)
plt.xlabel("Marks")
plt.ylabel("Number of students")
plt.title("Marks of the students of one group")
plt.show()
```

### Histogram

To make a histogram, we use the `.hist()` method of the `plt` module.

Some of the parameters of this method are:

* `x`: array or list of observations
* `bins`: to specify the size of the bins
* `range`: to specify the minimum and maximum of the bins.
* `histtype`: to indicate what type of histogram we want to draw
* `align`: to configure the alignment of the histogram bars
* `orientation`: to modify the orientation of the histogram (vertical or horizontal)
* `color`: to modify the color of the bars
* `edgecolor`: to modify the outline color of the bars 

```python
data = np.random.randint(200, size = 100)

plt.hist(data, bins = 10, range = (0, 200), color = "#ff85b0", edgecolor = "k")
plt.show()
```
